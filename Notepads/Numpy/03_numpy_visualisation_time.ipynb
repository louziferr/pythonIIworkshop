### Requirements

```python
import numpy as np

# for static plotting
import matplotlib.pyplot as plt

# for animation and dynamic output
import matplotlib.animation as animation
from IPython.display import HTML
```

# 4. NumPy: Visualisation time 

```python
# our x values
X = np.arange(1,101)
# our y values
Y = np.random.rand(100)*100

# you can either use plot
plt.plot(X,Y)
plt.show()
```

```python
# or scatter
plt.scatter(X,Y)
```

```python
X = np.arange(1,101)
# our y values
Y = np.random.rand(100)*100

fig, ax = plt.subplots()

def animate(time):
    ax.scatter(X[time], Y[time])

anim = animation.FuncAnimation(fig, animate, 50, interval=500)

HTML(anim.to_html5_video())
```

### Some mathematical functions

Numpy provides some mathematicalfunctions like 
    
                            cos(), sin(), exp()
                            
These functions are applied *elementwise* on numpy arrays:                     

```python
matrix = np.array([[3. , 2. , 6. ],
       [2. , 4. , 2.5],
       [3. , 8. , 4. ]])
# we take the Euler's number e as base an use each value as exponent
np.exp(matrix)
```

```python
X = np.array([1, 2, 3])
# we apply the sine function elementwise
np.sin(X)
```

#### Now we do the same plot as above, but with data from the sinus function

```python

X = np.arange(1,201)
Y = np.sin(X)

fig, ax = plt.subplots()

def animate(time):
    ax.scatter(X[time], Y[time], c = "orange")
    

anim = animation.FuncAnimation(fig, animate, 50, interval=500)

HTML(anim.to_html5_video())
```
